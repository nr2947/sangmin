```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
fx = sigmoid(x)

plt.grid()
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.ylim(-0.1, 1.1)
plt.plot(x, fx, label='sigmoid function')
plt.legend(loc='upper left')
plt.show()
```

```python
sigmoid_diff = fx * (1-fx)

plt.grid()
plt.xlabel('x')
plt.ylabel('sigmoid_derivative(x)')
plt.ylim(-0.1, 1.1)
plt.plot(x, sigmoid_diff, label='sigmoid derivative')
plt.legend()
plt.show()
```
